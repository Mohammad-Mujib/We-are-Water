### df 8 - depositos reguladores 2021

```python
depositos_reguladores_df = pd.read_excel('../data/Depositos reguladores.xlsx')
depositos_reguladores_df.head()
```

```python
# Drop the first row
depositos_reguladores_df = depositos_reguladores_df.drop([0], axis=0)
depositos_reguladores_df.head()
```

```python
# Reset the index
depositos_reguladores_df = depositos_reguladores_df.reset_index(drop=True)
depositos_reguladores_df.head()
```

```python
depositos_reguladores_df.columns
```

```python
# renaming the columns
depositos_reguladores_df.rename(columns={ '  Fecha de ': 'Fecha de construcción', 'Número de': 'Número de compartimentos', 
                                         '  Altura de': 'Altura de agua(m)', 'Capacidad': 'Capacidad(m³)'}, inplace=True)
depositos_reguladores_df.head()
```

```python
depositos_reguladores_df.drop('Superficie', axis=1, inplace=True)
```

```python
depositos_reguladores_df
```

```python
# saving the dataframe in csv
depositos_reguladores_df.to_csv('../cleaned-data/depositos_reguladores.csv')
```
