```python
import pandas as pd
```

### df 1 - datos embalses de 2021

```python
embalses_df = pd.read_excel('../data/Embalses.xlsx')
embalses_df.head()
```

```python
#dropping rows with null values
embalses_df = embalses_df.dropna()
```

```python
# Reset the index
embalses_df = embalses_df.reset_index(drop=True)
embalses_df.head()
```

```python
# renaming the columns
embalses_df.rename(columns={'Unnamed: 0': 'Denominación', 'Unnamed: 1': 'Río', 
                            'Unnamed: 2': 'Entrada en servicio', 'Unnamed: 3': 'Tipo',
                           'Unnamed: 4': 'Altura(m)', 'Unnamed: 5': 'Longitud(m)',
                           'Unnamed: 6': 'Capacidad(Hm3)', 'Unnamed: 7': 'Superficie(Ha)'}, inplace=True)
embalses_df
```

```python
embalses_df.info()
```

```python
# saving the dataframe in csv
embalses_df.to_csv('../cleaned-data/embalses.csv')
```

### df 2 - volumen de agua por embalses en Hm3(hectometros cubicos)

```python
volumen_embalses_df = pd.read_excel('../data/Volumen_agua_por_embalses.xls')
volumen_embalses_df.head()
```

```python
volumen_embalses_df = volumen_embalses_df.dropna()
```

```python
# to see all the columns
pd.options.display.max_columns = None
```

```python
# Reset the index
volumen_embalses_df = volumen_embalses_df.reset_index(drop=True)
volumen_embalses_df
# volumen in Hm3
```

```python
volumen_embalses_df.rename(columns={'Unnamed: 0': 'Embalses'}, inplace=True)
```

```python
volumen_embalses_df.head()
```

```python
# Create a dictionary to map the new column names
new_column_names = {
    f'Unnamed: {i}': str(1985 + i - 1) for i in range(1, 38)
}

# Rename the columns using the dictionary
volumen_embalses_df = volumen_embalses_df.rename(columns=new_column_names)
```

```python
volumen_embalses_df
```

```python
# Remove hyphens from the data in the DataFrame
volumen_embalses_df = volumen_embalses_df.replace('-', '0')
volumen_embalses_df
```

```python
# saving the dataframe in csv
volumen_embalses_df.to_csv('../cleaned-data/volumen_agua_por_embalses.csv')
```

### df 4 - consumo domestico por distritos de madrid en 2021

```python
consumo_distrito_df = pd.read_excel('../data/Consumo_doméstico_por_distrito.xlsx')
consumo_distrito_df.head()
# este data es de 2021
```

```python
consumo_distrito_df = consumo_distrito_df.drop(["Unnamed: 0", "Unnamed: 5"], axis=1)
```

```python
consumo_distrito_df = consumo_distrito_df.dropna()
consumo_distrito_df
```

```python
# Drop the first two rows
consumo_distrito_df = consumo_distrito_df.drop([3, 5], axis=0)
consumo_distrito_df
```

```python
# Reset the index
consumo_distrito_df = consumo_distrito_df.reset_index(drop=True)
consumo_distrito_df.head()
```

```python
# renaming the columns
consumo_distrito_df.rename(columns={'Unnamed: 1': 'Distrito', 'Unnamed: 2': 'Consumo(m3)', 
                            'Unnamed: 3': 'Habitantes', 'Unnamed: 4': 'Litros/habitante/día'}, inplace=True)
consumo_distrito_df.head()
```

```python
# Remove '01.' from the 'District' column
consumo_distrito_df['Distrito'] = consumo_distrito_df['Distrito'].str.replace(r'\d+\.', '')
```

```python
consumo_distrito_df
```

```python
# saving the dataframe in csv format
consumo_distrito_df.to_csv('../cleaned-data/consumo_doméstico_por_distrito.csv')
```

### df 6 - distribuciones por sectores en  miles de m3
